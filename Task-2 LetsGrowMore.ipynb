# Importing libraries

```python
import cv2
from IPython.display import display
from PIL import Image


```

# Loading the Image

```python
image = cv2.imread("dog image.jpg")
image = cv2.resize(image,(600,400))
display(Image.fromarray(image))

```

# Converting into Gray Scale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

# Inverting the Image

```python
inverted_image = 255 - gray_image
display(Image.fromarray(inverted_image))
```

# Blurring the Image Using Gaussian Blur

```python
blurred = cv2.GaussianBlur(inverted_image, (5, 5), 0)
display(Image.fromarray(blurred))
```

# Inverting the Image to get the Pencil Sketch

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
display(Image.fromarray(pencil_sketch))
```

# Comparing the Original Image with Pencil Sketch

```python
display(Image.fromarray(image))
display(Image.fromarray(pencil_sketch))
```
